##Find the names of the unique launch sites

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('spacex_comprehensive_launches.csv')

# Inspect 'Date' column
print("Info about 'Date' column:")
print(df['Date'].head())
print(df['Date'].dtype)

# Inspect 'Launchoutcome' column
print("\nUnique values in 'Launchoutcome' column:")
print(df['Launchoutcome'].unique())
print(df['Launchoutcome'].value_counts(dropna=False).to_markdown(numalign="left", stralign="left"))

# Inspect 'Outcome' column
print("\nUnique values in 'Outcome' column:")
print(df['Outcome'].unique())
print(df['Outcome'].value_counts(dropna=False).to_markdown(numalign="left", stralign="left"))

# Convert 'Date' to datetime and extract year
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Assume 'Outcome' is the most reliable indicator of success (True for success, False for failure)
# Convert 'Outcome' to boolean, coercing errors (NaN will become False)
df['Success'] = df['Outcome'].astype(bool)

# Calculate yearly average success rate
yearly_success_rate = df.groupby('Year')['Success'].mean().reset_index()
yearly_success_rate.rename(columns={'Success': 'Average Success Rate'}, inplace=True)

# Create the line chart
plt.figure(figsize=(12, 7))
plt.plot(yearly_success_rate['Year'], yearly_success_rate['Average Success Rate'], marker='o')
plt.xlabel('Year')
plt.ylabel('Average Success Rate')
plt.title('Yearly Average Success Rate')
plt.grid(True)
plt.xticks(yearly_success_rate['Year']) # Ensure all years are shown as x-ticks
plt.ylim(0, 1) # Set y-axis limits from 0 to 1 for rate
plt.tight_layout()

# Save the plot
plt.savefig('yearly_average_success_rate_line_chart.png')

print("Line chart of yearly average success rate generated successfully: yearly_average_success_rate_line_chart.png")
```

# success rate of each orbit type

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('spacex_comprehensive_launches.csv')

# Convert 'Outcome' to numeric for success rate calculation (True=1, False=0, NaN will be ignored by mean)
df['Outcome_Numeric'] = df['Outcome'].astype(float)

# Calculate success rate for each orbit type
# Group by 'Orbit' and calculate the mean of 'Outcome_Numeric'
# Drop rows where 'Orbit' is NaN before grouping if desired, or handle 'nan' in the orbit types
# Let's convert 'Orbit' to string to handle 'nan' and ensure consistency in grouping
df['Orbit'] = df['Orbit'].astype(str)

success_rate_by_orbit = df.groupby('Orbit')['Outcome_Numeric'].mean().reset_index()
success_rate_by_orbit.rename(columns={'Outcome_Numeric': 'Success Rate'}, inplace=True)

# Sort by Success Rate in descending order
success_rate_by_orbit = success_rate_by_orbit.sort_values(by='Success Rate', ascending=False)

# Create the bar chart
plt.figure(figsize=(12, 7))
plt.bar(success_rate_by_orbit['Orbit'], success_rate_by_orbit['Success Rate'], color='skyblue')
plt.xlabel('Orbit Type')
plt.ylabel('Success Rate')
plt.title('Success Rate of Each Orbit Type')
plt.ylim(0, 1) # Success rate is between 0 and 1
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()

# Save the plot
plt.savefig('success_rate_by_orbit_bar_chart.png')

print("Bar chart of success rate by orbit type generated successfully: success_rate_by_orbit_bar_chart.png")
```

#Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('spacex_comprehensive_launches.csv')

# Convert 'Date' to datetime objects and remove timezone information
df['Date'] = pd.to_datetime(df['Date']).dt.tz_localize(None)

# Define the date range (ensure these are timezone-naive as well)
start_date = pd.to_datetime('2010-06-04')
end_date = pd.to_datetime('2017-03-20')

# Filter DataFrame by date range
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Define drone ship and ground pad names
drone_ship_names = ['Of Course I Still Love You', 'Just Read The Instructions', 'A Shortfall of Gravitas']
ground_pads_names = ['Landing Zone 1', 'Landing Zone 4']

# Function to categorize landing outcomes
def categorize_landing_outcome(row):
    booster_landing = row['Booster landing']
    landing_pad = str(row['LandingPad']) # Convert to string to handle nan

    if booster_landing == 'Success':
        if landing_pad in drone_ship_names:
            return 'Success (drone ship)'
        elif landing_pad in ground_pads_names:
            return 'Success (ground pad)'
        else:
            return 'Success (other pad)'
    elif booster_landing == 'Failure':
        if landing_pad in drone_ship_names:
            return 'Failure (drone ship)'
        elif landing_pad in ground_pads_names:
            return 'Failure (ground pad)'
        else:
            return 'Failure (other pad)'
    elif booster_landing == 'No Attempt':
        return 'No Attempt'
    else:
        return 'Unknown'

# Apply the categorization function to create a new column
filtered_df['LandingOutcomeCategory'] = filtered_df.apply(categorize_landing_outcome, axis=1)

# Calculate the count of each category and rank in descending order
landing_outcome_counts = filtered_df['LandingOutcomeCategory'].value_counts().sort_values(ascending=False)

if not landing_outcome_counts.empty:
    print("Rank of landing outcomes between 2010-06-04 and 2017-03-20:")
    print(landing_outcome_counts.to_markdown(numalign="left", stralign="left"))
else:
    print("No landing outcomes found within the specified date range.")
```

# ist the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('spacex_comprehensive_launches.csv')

# Convert 'Date' to datetime and extract Year
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Define drone ship names
drone_ship_names = ['Of Course I Still Love You', 'Just Read The Instructions', 'A Shortfall of Gravitas']

# Filter for year 2015, failed landing outcomes, and drone ship landings
failed_drone_ship_landings_2015 = df[
    (df['Year'] == 2015) &
    (df['Booster landing'] == 'Failure') &
    (df['LandingPad'].isin(drone_ship_names))
]

# Select the required columns
result = failed_drone_ship_landings_2015[['Booster landing', 'BoosterVersion', 'LaunchSite']]

if not result.empty:
    print("Failed landing outcomes in drone ship, their booster versions, and launch site names for 2015:")
    print(result.to_markdown(index=False, numalign="left", stralign="left"))
else:
    print("No failed drone ship landings found in 2015 matching the criteria.")
```

##List the names of the booster which have carried the maximum payload mass

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('spacex_comprehensive_launches.csv')

# Convert 'PayloadMass' to numeric, coercing errors to NaN
df['PayloadMass'] = pd.to_numeric(df['PayloadMass'], errors='coerce')

# Find the maximum payload mass
max_payload_mass = df['PayloadMass'].max()

# Filter the DataFrame to get records with the maximum payload mass
boosters_with_max_payload = df[df['PayloadMass'] == max_payload_mass]

# Get the unique names of the boosters
booster_names = boosters_with_max_payload['BoosterVersion'].unique().tolist()

if booster_names:
    print(f"The booster(s) which have carried the maximum payload mass ({max_payload_mass}) are:")
    for booster in booster_names:
        print(f"- {booster}")
else:
    print("No boosters found with a maximum payload mass in the dataset.")
```
